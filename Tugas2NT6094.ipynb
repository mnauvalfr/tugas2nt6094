```python
import numpy as np
import matplotlib.pyplot as plt
from skimage.io import imread, imshow
from skimage.color import rgb2hsv, rgb2gray, rgb2yuv
from skimage import color, exposure, transform
from skimage.exposure import equalize_hist
```

Citra gambar yang diolah FFT adlaah zeolit. Zeolit adalah suatu lempung yang banyak diteliti karena sifatnya yang mudah menyerap bahan dan berpori, jadi banyak diteliti sebagai katalis.

Pada tugas ini, diolah data citra zeolit raw dan yang dilakukan treatment. 
Pengolahan data menggunakan FFT dilakukan untuk menganalisis defek yang terjadi akibat treatment tersebut.

Data Zeolit 1

```python
dark_image = imread('zeolitA.jpg')
```

```python
dark_image_grey = rgb2gray(dark_image)
plt.figure(num=None, figsize=(8, 6), dpi=80)
plt.imshow(dark_image_grey, cmap='gray');
```

FFT Zeolit 1

```python
dark_image_grey_fourier = np.fft.fftshift(np.fft.fft2(dark_image_grey))
plt.figure(num=None, figsize=(8, 6), dpi=80)
plt.imshow(np.log(abs(dark_image_grey_fourier)), cmap='gray');
```

Data Zeolit 2

```python
dark_image2 = imread('zeolitB.jpg')
dark_image_grey1 = rgb2gray(dark_image2)
plt.figure(num=None, figsize=(8, 6), dpi=80)
plt.imshow(dark_image_grey1, cmap='gray');
```

FFT Zeolit 2

```python
dark_image_grey_fourier2 = np.fft.fftshift(np.fft.fft2(dark_image_grey1))
plt.figure(num=None, figsize=(8, 6), dpi=80)
plt.imshow(np.log(abs(dark_image_grey_fourier2)), cmap='gray');
```

Dengan membandingkan hasil FFT, dapat dilihat secara semi-kuantitatif perubahan struktur akibat treatment. Terlihat bahwa treatment menghasilkan pola FFT yang berbentuk relatif sama, namun jauh lebih nyata (puncaknya jauh lebih terlihat). Morflogi zeolit, sebelum treatment, adalah kubus (cubic). Sesudah dilakukan treatment, secara morfologi struktur terlihat terabrasi (amorf) Namun, pola yang relatif sama ini mengimplikasikan bahwa struktur zeolit relatif tetap, dengan treatment 'memotong' material zeolit menjadi banyak lapisan-lapisan yang bentuknya persegi, dan jika dilihat dari jarak jauh terlihat 'kasar' atau 'bergerigi'.

Pengolahan data ini dapat dilakukan secara kuantitatif pula. Perbandingan defek ini dapat dijadikan bahan penulisan jurnal yang mengurusi bahan berpori.
